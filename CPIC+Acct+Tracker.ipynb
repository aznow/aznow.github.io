```python
print('Importing libraries...')
# =========== importing libraries ===========
import pandas as pd
import plotly
from plotly.offline import plot
import plotly.graph_objs as go
from plotly.graph_objs import Scatter, Layout
import plotly.figure_factory as ff
from pandas import ExcelWriter
from pandas import ExcelFile
from IPython.core.display import display, HTML
# =========== begin algs ===========
print('Done')
plotly.offline.init_notebook_mode(connected=True)  # initializes offline plotting
# ============= Excel import ================
xl_file = pd.ExcelFile("cpic_now_acct.xlsx")
# print(xl_file.sheet_names[0])
# df = pd.read_excel("cpic_now_acct.xlsx")
df = xl_file.parse(xl_file.sheet_names[0])
df = df.fillna('')

df2 = xl_file.parse(xl_file.sheet_names[1])
df2 = df2.fillna('')

df3 = xl_file.parse(xl_file.sheet_names[2])
df3 = df3.fillna('')

df4 = xl_file.parse(xl_file.sheet_names[3])
df4 = df4.fillna('')

def color_cols(x): # color function
    # copy df to new - original data is not changed
    df = x.copy()
    # select all values to default value - no color (necessary for some reason)
    df.loc[:,:] = 'background-color: none'
    # overwrite default values w/ desired colors      
    df[['credit']] = 'background-color: #e5f9de'
    df[['debit']] = 'background-color: #f9dfde'
#     df.loc[35:36,1:3] = 'font-weight: bold; background-color: none'
#     df.loc[36,3:4] = 'font-weight: bold'
#     df.loc[27:29,:] = 'background-color: none'
#     df.loc[29,1:2] = 'font-weight: bold'
    # return colored df
    return df

def color_cols2(x): # color function
    # copy df to new - original data is not changed
    df = x.copy()
    # select all values to default value - no color (necessary for some reason)
    df.loc[:,:] = 'background-color: none'
    # overwrite default values w/ desired colors
    df.loc[13, 0:3] = 'font-weight: bold; background-color: none' # row:row, col:col
    df.loc[15:18, 0:3] = 'font-weight: bold'
    df[['Tot Credits']] = 'background-color: #e5f9de'
    df[['Tot Debits']] = 'background-color: #f9dfde'
    df.loc[8:12,:] = 'background-color: none'
#     df.loc[29,1:2] = 'font-weight: bold'
    # return colored df
    return df    

# =========== Create plot ===================
print ('Building plot...')
#df = df.style.applymap(lambda v: 'background-color: %s' % 'green' if v=='col' else '')
# df_list1 = df['date'].tolist()  # extracts column as python list
# df_list2 = df['balance'].tolist()  # extracts column as list

df_list1 = df.loc[2:33,'date']  # extracts column as python list
df_list2 = df.loc[2:33,'balance']  # extracts column as python list

df_list3 = df2.loc[2:10,'date']  # extracts column as python list
df_list4 = df2.loc[2:10,'balance']  # extracts column as python list

df_list5 = df3.loc[0:5,'Date'].tolist()  # extracts column as python list
df_list6 = df3.loc[0:5,'Balance'].tolist()  # extracts column as python list

# Create traces
trace1 = go.Scatter(
    x = df_list1,
    y = df_list2,
    mode = 'lines+markers',
    name = 'DS Checking',
    line = dict(
        color = ('#9c43a8'),
        width = 2)
)

trace2 = go.Scatter(
    x = df_list3,
    y = df_list4,
    mode = 'lines+markers',
    name = 'DS Savings',
    line = dict(
        color = ('#e7a7ef'),
        width = 2)
)

trace3 = go.Scatter(
    x = df_list5,
    y = df_list6,
    mode = 'lines+markers',
    name = 'Paypal',
    line = dict(
        color = ('#f47d42'),
        width = 2)
)

data = [trace1, trace2, trace3]

layout = dict(title = 'Balance History',
              yaxis = dict(title= 'Balance ($)', zeroline = True),
              xaxis = dict(title= 'Date-time', zeroline = True)
             )

fig = dict(data=data, layout=layout)

print ('Done')
df = df.style.apply(color_cols, axis=None)  # applies color function to dataframe
df2 = df2.style.apply(color_cols, axis=None)
df4 = df4.style.apply(color_cols2, axis=None)

display(HTML('<div class="intro" style="width: 100%;"><a href="https://centralphoenixnow.org"><img style="position: absolute; left:45%; top: 10px; z-index: 0;" class="logo" src="https://aznow.github.io/Logo_Color_2x.png" alt="CPIC &#8226; Logo" width="150" height="150"></a><a href="https://aznow.github.io/cpic_tracker_qrs.png"><img style="position: absolute; left:85%; top: 10px; z-index: 0;" class="logo" src="https://aznow.github.io/cpic_tracker_qrs.png" alt="QRS Code" width="150" height="150"></a><br><br><br></div><div class="line" style="width: 100%; height: 2px; position: absolute; left: 0px; top: 290px; background-color:#9c43a8;"></div>') )
             
             
plotly.offline.iplot(fig, filename='cpic_acct_data1.html')
display(HTML('<h3>DS Checking</h3>') )
display(df) #  displays dataframe

display(HTML('<h3>DS Savings</h3>') )
display(df2) #  displays dataframe

display(HTML('<h3>Paypal</h3>') )
df3 = df3.drop(['Name'], axis=1)
display(df3) #  displays dataframe

display(HTML('<h3>Totals</h3>') )
display(df4) #  displays dataframe
```
